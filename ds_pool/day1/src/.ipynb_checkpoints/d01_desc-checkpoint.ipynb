```python
import calendar
try:
    number_of_month =     int(input("month: "))           # глобальная переменная месяца даты рождения, необходимо присвоение значения
    number_of_day =       int(input("day: "))            # глобальная переменная дня даты рождения, необходимо присвоение значения
except ValueError:
    print ("input isnt num")
else:
    def month_of_birth(month, m_o_b = number_of_month):
        "Функция принимает значение month, сравнивает с m_o_b и возвращает результаты сравнения"

        if (month == m_o_b):
            return ("да")
        if (month > m_o_b):
            return ("раньше")
        if (month < m_o_b):
            return ("позже")

    def day_of_birth(day, d_o_b = number_of_day):
        "Функция принимает значение day, сравнивает его с d_o_b и возвращает результаты сравнения"

        if (day == d_o_b):
            return ("да")
        if (day > d_o_b):
            return ("раньше")
        if (day < d_o_b):
            return ("позже")

    def get_month():
        """Функция формирует запросы month и анализирует результаты сравнения, переводит номер найденного месяца в 
        строку его названия и возвращает результат"""
        month = 6
        start = 0
        end = 13
        while (month > 0 and month < 13):
            if (month_of_birth(month).lower().replace(' ', '') == "да"):
                return (month)
            if (month_of_birth(month).lower().replace(' ', '') == "раньше"):
                end, month = month, ((month + start) // 2)
            if (month_of_birth(month).lower().replace(' ', '') == "позже"):
                start, month = month, ((month + end) // 2 + ((month + end) % 2))
        return (-1)


    def get_day():
        "Функция формирует запросы day и анализирует результаты сравнения, возвращает результат"

        day = 16
        start = 0
        end = 32
        while (day > 0 and day < 32):
            if (day_of_birth(day).lower().replace(' ', '') == "да"):
                return (day)
            if (day_of_birth(day).lower().replace(' ', '') == "раньше"):
                end, day = day, ((day + start) // 2)
            if (day_of_birth(day).lower().replace(' ', '') == "позже"):
                start, day = day, ((day + end) // 2 + ((day + end) % 2))
        return (-1)

    def birthday():                    # функция, запускающая процесс определения даты рождения и выводящая результат
        mon = get_month()
        day = get_day()
        if (mon == -1 or day == -1):
            print ("wrong input")
        elif (mon == 2 and day > 29):
            print("non-existent day in feb")
        elif (mon in [4, 6, 9, 11] and day > 30):
                print("non-existent day")
        else:
            print("month of your birth: {}, day: {}".format(calendar.month_name[mon], day))
    
    birthday()
```
