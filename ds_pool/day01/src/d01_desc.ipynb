```python
from datetime import datetime, timedelta, date, time
"""Решение задания 2"""

def input_date():
    str_with_date_start = input("введите дату начала работ в формате gggg-mm-dd: ")
    str_with_date_end = input("введите дату окончания работ в формате gggg-mm-dd: ")
    #str_with_date_start = "2022-4-3"
    #str_with_date_end = "2022-4-18"
    
    return (str_with_date_start, str_with_date_end)

def date_to_datetime(start_date, end_date):
    start_date = datetime.strptime(start_date, '%Y-%m-%d')
    end_date = datetime.strptime(end_date, '%Y-%m-%d')
    end_date = end_date + timedelta(days=1) 
    
    return (start_date, end_date)

def delta_time(start_date, end_date):
    return ((end_date - start_date)).days

def day_of_the_week(start_date):
    return (datetime.isoweekday(start_date))
    

def calculating_hours(start_day_of_the_week, delta):
    work_hours = ((delta // 7) * 40)
    bonus = delta % 7
    while(bonus > 0):
        if (start_day_of_the_week != 6 and start_day_of_the_week != 7):
            work_hours += 8
            start_day_of_the_week += 1
        else:
            if (start_day_of_the_week == 6):
                start_day_of_the_week += 1
            else:
                start_day_of_the_week = 1
        bonus -= 1
    return (work_hours)
```

```python
str_start, str_end = input_date()   
while (str_start and str_end):
    try:
        date_start, date_end = date_to_datetime(str_start, str_end)
    except ValueError:
        print("bad input")
    except TypeError:
        print("bad input")    
    else:
        if (date_end < date_start):
            print("end < start, bad")
        else:
            print("всего рабочих часов: ", calculating_hours(day_of_the_week(date_start), delta_time(date_start, date_end)))
    finally:
        str_start, str_end = input_date()
```

```python
def get_all_info(str):   
    date_as_dt = datetime.strptime(str, '%Y-%m-%d %H:%M:%S')
    dad = datetime.date(date_as_dt)
    tat = datetime.time(date_as_dt)
    n_week = datetime.date(date_as_dt).isocalendar()[1]
    n_day_name = calendar.day_name[datetime.isoweekday(date_as_dt) - 1]
    all_sec = tat.second + tat.minute * 60 + tat.hour * 3600
    
    
    info = {'Дата': dad, 'Год': dad.year,'Месяц': dad.month,'День': dad.day,
            'номер_недели': n_week, 'День_недели': n_day_name,
            'Время': tat, 'Часы': tat.hour, 'Минуты': tat.minute, 'Секунды': tat.second,
            'секунд_с_начала_дня': all_sec}

    return info

    
df = pd.read_excel('../datasets/Даты.xlsx')
new_cols = ['Дата','Год','Месяц','День','номер_недели', 'День_недели', 'Время', 'Часы', 'Минуты', 'Секунды', 'секунд_с_начала_дня']
for col in new_cols:
    df[col] = ""
i = 0
for str in df.loc[:,'datetime']:
    infa = get_all_info(str)
    for col_name in new_cols:
        df[col_name][i] = infa[col_name]
    
    i += 1
print(df)
nik_name = 'pskip'
df.to_excel('../datasets/Даты '+nik_name+'.xlsx', index=False)

```
